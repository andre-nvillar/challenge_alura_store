### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Calcular o faturamento (soma da coluna 'preco') de cada loja
faturamento1 = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()

# Organizar os resultados em um dicionário
faturamentos = {
    "Loja 1": faturamento1,
    "Loja 2": faturamento2,
    "Loja 3": faturamento3,
    "Loja 4": faturamento4
}
```

```python
# Transformar em DataFrame para melhor visualização
df_faturamento = pd.DataFrame.from_dict(faturamentos, orient='index', columns=['Faturamento Total'])
df_faturamento = df_faturamento.sort_values(by='Faturamento Total', ascending=False)

print("Faturamento total de cada loja:")
display(df_faturamento)
```

# 2. Vendas por Categoria


```python
# Loja 1
categorias_loja1 = loja.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
print("Loja 1 - Produtos vendidos por categoria:")
display(categorias_loja1)

# Loja 2
categorias_loja2 = loja2.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
print("Loja 2 - Produtos vendidos por categoria:")
display(categorias_loja2)

# Loja 3
categorias_loja3 = loja3.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
print("Loja 3 - Produtos vendidos por categoria:")
display(categorias_loja3)

# Loja 4
categorias_loja4 = loja4.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
print("Loja 4 - Produtos vendidos por categoria:")
display(categorias_loja4)
```

# 3. Média de Avaliação das Lojas

```python
# Média de avaliação dos clientes por loja
avaliacao1 = loja['Avaliação da compra'].mean()
avaliacao2 = loja2['Avaliação da compra'].mean()
avaliacao3 = loja3['Avaliação da compra'].mean()
avaliacao4 = loja4['Avaliação da compra'].mean()

# Organizar em dicionário
avaliacoes = {
    "Loja 1": avaliacao1,
    "Loja 2": avaliacao2,
    "Loja 3": avaliacao3,
    "Loja 4": avaliacao4
}

# Converter em DataFrame para visualizar melhor
df_avaliacoes = pd.DataFrame.from_dict(avaliacoes, orient='index', columns=['Média Avaliações'])
df_avaliacoes = df_avaliacoes.sort_values(by='Média Avaliações', ascending=False)

print("Média das avaliações dos clientes por loja:")
display(df_avaliacoes)
```

# 4. Produtos Mais e Menos Vendidos

```python
# Função para analisar produtos mais e menos vendidos em uma loja
def analise_produtos(df, nome_loja):
    # Contagem de vendas por produto
    produtos = df.groupby('Produto')['Produto'].count().sort_values(ascending=False)

    print(f"\n🔹 {nome_loja} - Produtos mais vendidos:")
    display(produtos.head(5))  # Top 5 mais vendidos

    print(f"\n🔸 {nome_loja} - Produtos menos vendidos:")
    display(produtos.tail(5))  # 5 menos vendidos

    # Gráfico - Top 5 mais vendidos
    plt.figure(figsize=(8,5))
    produtos.head(5).plot(kind='bar', color='green')
    plt.title(f"{nome_loja} - Top 5 Produtos Mais Vendidos")
    plt.ylabel("Quantidade de Vendas")
    plt.xlabel("Produto")
    plt.xticks(rotation=45)
    plt.show()

    # Gráfico - Top 5 menos vendidos
    plt.figure(figsize=(8,5))
    produtos.tail(5).plot(kind='bar', color='red')
    plt.title(f"{nome_loja} - Top 5 Produtos Menos Vendidos")
    plt.ylabel("Quantidade de Vendas")
    plt.xlabel("Produto")
    plt.xticks(rotation=45)
    plt.show()

```

```python
# Executando para cada loja
analise_produtos(loja, "Loja 1")
analise_produtos(loja2, "Loja 2")
analise_produtos(loja3, "Loja 3")
analise_produtos(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
# Cálculo do custo médio de frete por loja
frete_medio_lojas = {
    "- Loja 1": loja["Frete"].mean(),
    "- Loja 2": loja2["Frete"].mean(),
    "- Loja 3": loja3["Frete"].mean(),
    "- Loja 4": loja4["Frete"].mean()
}

# Exibindo os resultados
print("Custo Médio de Frete por Loja:\n")
for nome, frete in frete_medio_lojas.items():
    print(f"{nome}: R$ {frete:.2f}")
```

```python
# Plotando o gráfico do frete
plt.figure(figsize=(8,5))
plt.bar(frete_medio_lojas.keys(), frete_medio_lojas.values(), color="skyblue", edgecolor="black")
plt.title("Custo Médio de Frete por Loja")
plt.ylabel("Custo Médio (R$)")
plt.xlabel("Lojas")
plt.show()
```

# 6 - Gráficos



*   Faturamento total e média de avaliação por loja


```python
# Acessando os dados
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Dicionário com os dados de cada loja
lojas_df = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4,
}

# --- Gráfico 1: Faturamento e Avaliação Média por Loja ---
faturamentos = {nome: df['Preço'].sum() for nome, df in lojas_df.items()}
avaliacoes = {nome: df['Avaliação da compra'].mean() for nome, df in lojas_df.items()}

# Criando o DataFrame para o gráfico
df_analise = pd.DataFrame({
    'Faturamento': faturamentos,
    'Avaliação Média': avaliacoes
})

fig, ax1 = plt.subplots(figsize=(12, 7))

# Gráfico de Barras para Faturamento
color = 'tab:blue'
ax1.set_xlabel('Loja')
ax1.set_ylabel('Faturamento (R$)', color=color)
ax1.bar(df_analise.index, df_analise['Faturamento'], color=color, alpha=0.6, label='Faturamento')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_title('Faturamento Total e Média de Avaliação por Loja', fontsize=16)

# Eixo Y secundário para a Avaliação
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Avaliação Média', color=color)
ax2.plot(df_analise.index, df_analise['Avaliação Média'], color=color, marker='o', linestyle='--', label='Avaliação Média')
ax2.tick_params(axis='y', labelcolor=color)

# Linhas de referência para a média total
faturamento_medio_geral = df_analise['Faturamento'].mean()
avaliacao_media_geral = df_analise['Avaliação Média'].mean()
ax1.axhline(faturamento_medio_geral, color='blue', linestyle=':', linewidth=2, label='Média Geral Faturamento')
ax2.axhline(avaliacao_media_geral, color='red', linestyle=':', linewidth=2, label='Média Geral Avaliação')

# Legendas combinadas
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

fig.tight_layout()
plt.show()
```



*   Relação entre preço do produto custo do frete entre todas as lojas



```python
# Concatenando todos os dataframes para uma análise conjunta
df_total = pd.concat(lojas_df.values(), ignore_index=True)

plt.figure(figsize=(12, 7))
plt.scatter(df_total['Preço'], df_total['Frete'], alpha=0.5, color='purple', s=df_total['Avaliação da compra'] * 20)
plt.title('Relação entre Preço do Produto e Custo do Frete', fontsize=16)
plt.xlabel('Preço do Produto (R$)', fontsize=12)
plt.ylabel('Custo do Frete (R$)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Adicionando um texto explicativo
plt.text(
    0.88,
    0.50,
    'Tamanho do círculo = Avaliação do produto',
    horizontalalignment='right',
    verticalalignment='top',
    transform=plt.gca().transAxes,
    fontsize=10,
    bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8)
)

plt.show()
```

* Produtos mais e menos vendidos;

* Conclusão parcial: A Loja 1 é mais sólida pela diversificação, enquanto a Loja 4 carece de força em qualquer segmento.

*   Análise de desempenho geográfico



```python
# Carregando os dados e adicionando uma coluna 'loja' para identificar a origem
loja1 = pd.read_csv(url)
loja1['loja'] = 'Loja 1'

loja2 = pd.read_csv(url2)
loja2['loja'] = 'Loja 2'

loja3 = pd.read_csv(url3)
loja3['loja'] = 'Loja 3'

loja4 = pd.read_csv(url4)
loja4['loja'] = 'Loja 4'

# Concatenando todos os dataframes em um só
df_total = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Verificando as primeiras linhas para confirmar a estrutura
print(df_total.head())
```

Gráfico de Dispersão para as 4 lojas

```python
plt.figure(figsize=(12, 10))

# Mapeando os dados de cada loja
cores = {'Loja 1': 'blue', 'Loja 2': 'green', 'Loja 3': 'red', 'Loja 4': 'purple'}

for loja, cor in cores.items():
    df_loja = df_total[df_total['loja'] == loja]
    plt.scatter(
        df_loja['lon'],
        df_loja['lat'],
        c=cor,
        label=loja,
        alpha=0.6,
        s=100  # Tamanho do ponto
    )

plt.title('Distribuição Geográfica de Vendas por Loja', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Conclusões:
# - A maioria das vendas está concentrada em uma área central, com forte sobreposição entre as lojas.
# - As lojas 1 e 2 parecem ter uma distribuição de vendas mais ampla do que as lojas 3 e 4.
# - Não há uma segmentação geográfica clara entre as lojas; elas competem na mesma região.
```

Gráfico de Dispersão de Vendas por Faturamento

```python
plt.figure(figsize=(12, 10))

# A cor e o tamanho do ponto serão baseados no preço do produto
scatter = plt.scatter(
    df_total['lon'],
    df_total['lat'],
    c=df_total['Preço'],
    cmap='viridis',  # Mapa de cores para o preço
    s=df_total['Preço'] / 5,  # O tamanho do ponto é proporcional ao preço
    alpha=0.6
)

plt.colorbar(scatter, label='Preço do Produto (R$)')
plt.title('Distribuição Geográfica das Vendas (por Preço do Produto)', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Conclusões:
# - As vendas com os maiores preços (pontos amarelos e maiores) parecem ocorrer em toda a área de cobertura, sem uma concentração específica em uma única região.
# - Não há um padrão geográfico que indique que vendas de alto valor estejam restritas a uma latitude/longitude específica.
```

Avaliações e frete por região

```python
# Criando quadrantes geográficos para análise
# Neste exemplo, vamos dividir em 4 quadrantes para simplificar
lat_media = df_total['lat'].median()
lon_media = df_total['lon'].median()

def get_quadrante(row):
    if row['lat'] >= lat_media and row['lon'] >= lon_media:
        return 'Nordeste'
    elif row['lat'] >= lat_media and row['lon'] < lon_media:
        return 'Noroeste'
    elif row['lat'] < lat_media and row['lon'] >= lon_media:
        return 'Sudeste'
    else:
        return 'Sudoeste'

df_total['quadrante'] = df_total.apply(get_quadrante, axis=1)

# Calculando as médias de frete e avaliação por quadrante
analise_regiao = df_total.groupby('quadrante').agg({
    'Frete': 'mean',
    'Avaliação da compra': 'mean',
    'Preço': 'sum'
}).rename(columns={'Preço': 'Faturamento'})

print('Análise de Desempenho por Região Geográfica:')
display(analise_regiao.sort_values(by='Faturamento', ascending=False))

# Conclusões:
# - O quadrante "Noroeste" tem o maior faturamento, a melhor avaliação média e o menor frete médio.
# - O quadrante "Sudeste" tem o pior desempenho nas três métricas.
# - Isso sugere que a localização geográfica dos clientes, e não a loja, pode ser um fator determinante para o sucesso das vendas e a satisfação do cliente.
```
